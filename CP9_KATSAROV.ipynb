```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Particle gun: Electrons with energy 300 MeV

### Lead shield of 2cm thickness and water sphere of 50cm radius

```python
shield = pd.read_csv("output_nt_Shield.csv")
water = pd.read_csv("output_nt_Water.csv")
```

```python
shield_energies = shield[4:].values.astype("float")
```

```python
water_energies = water[4:].values.astype("float")
```

```python
print("Total events: ",shield_energies.shape[0])
```

```python
plt.hist(shield_energies,bins=np.linspace(50,300,50))
plt.title("Shield deposited energy")
plt.xlabel("Deposited energy (MeV)")
plt.ylabel("Counts")
plt.show()
```

```python
plt.hist(water_energies,bins=np.linspace(0,200,50))
plt.title("Water deposited energy")
plt.xlabel("Deposited energy (MeV)")
plt.ylabel("Counts")
plt.show()
```

```python
#ratio of deposited energies
ratio = np.average(water_energies/shield_energies)
```

```python
print("Average ratio of energy deposited in the two objects: ", round(ratio,5))
```

### Changing the material and thickness of shield

### Tungsten shield of 1.5cm thickness and water sphere of 50cm radius

```python
shield2 = pd.read_csv("output_nt_Shield2.csv")
water2 = pd.read_csv("output_nt_Water2.csv")
```

```python
shield_energies2 = shield2[4:].values.astype("float")
```

```python
water_energies2 = water2[4:].values.astype("float")
```

```python
print("Total events: ",shield_energies2.shape[0])
```

```python
plt.hist(water_energies2,bins=np.linspace(0,200,50))
plt.title("Water deposited energy")
plt.xlabel("Deposited energy (MeV)")
plt.ylabel("Counts")
plt.show()
```

The average ratio of deposited energy decreases.

We can see that the slightly thinner tungsten shield actually does better than the thicker lead shield.

This is because tungsten is a better absorber of radiation than lead and requires less thickness to provide similar levels of protection.
